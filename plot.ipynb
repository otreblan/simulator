```python
from dask.distributed import Client

import matplotlib.pyplot as plt
import pandas as pd
import glob
```

```python
client = Client(threads_per_worker=1)

client
```

```python
def plot_csv(csv_path: str):
    new_path = csv_path.replace(".csv", ".png")
    df = pd.read_csv(csv_path).set_index("x")
    figure = df.plot(ylim=(-1, 1)).get_figure()
    figure.savefig(new_path, dpi=300)
    plt.close(figure)
```

```python
futures = client.map(plot_csv, glob.glob("build/sim*.csv"))

result = client.gather(futures)
```
